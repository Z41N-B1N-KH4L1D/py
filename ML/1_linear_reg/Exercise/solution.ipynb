```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (in USD)")
plt.scatter(df["year"], df["per capita income (US$)"], marker='x', color='red')
```

```python
model = linear_model.LinearRegression()
model.fit(df[["year"]], df[["per capita income (US$)"]])
```

```python
model.predict([[2020]])
```

### Following code is just to see the regression line.

```python
# Plot data with regression line and show prediction for 2020
plt.scatter(df['year'], df['per capita income (US$)'], marker='x', color='red')
years = np.array(df['year']).reshape(-1,1)
preds = model.predict(years)
plt.plot(df['year'], preds, color='blue')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income with Linear Regression')
plt.show()
print('Predicted income for 2020:', float(model.predict([[2020]])))
```
